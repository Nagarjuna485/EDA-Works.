# Im-Balanced Data Classification

## Importing Dataset

```python
from sklearn.datasets import make_classification

X, y = make_classification(n_samples=10000, n_features=2, n_redundant=0, 
                           n_clusters_per_class=1, weights=[0.90], flip_y=0, random_state=1)
```

```python
X.shape
```

```python
y.shape
```

```python
from collections import Counter
print('Original dataset shape {}'.format(Counter(y)))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.countplot(y)
plt.grid()
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(X[:,0], X[:,1], c=y)
plt.show()
```

### Over Samping With SMOTE

```python
from imblearn.over_sampling import SMOTE
smt = SMOTE(random_state=0)
X_smt, y_smt = smt.fit_sample(X, y)
```

```python
X_smt.shape
```

```python
y_smt.shape
```

```python
print('Smote Applying y_smt dataset shape {}'.format(Counter(y_smt)))
```

```python
sns.countplot(y_smt)
plt.grid()
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(X_smt[:,0], X_smt[:,1], c=y_smt)
plt.show()
```

### Under Sampling Random Under Sampler 

```python
from imblearn.under_sampling import RandomUnderSampler 

rus = RandomUnderSampler(random_state=42)
X_rus, y_rus = rus.fit_sample(X, y)
```

```python
X_rus.shape
```

```python
y_rus.shape
```

```python
print('Under Sampling y_rus shape {}'.format(Counter(y_rus)))
```

```python
sns.countplot(y_rus)
plt.grid()
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(X_rus[:,0], X_rus[:,1], c=y_rus)
plt.show()
```
